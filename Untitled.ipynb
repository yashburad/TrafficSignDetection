```python
import skimage
from PIL import Image
import os
import cv2
import skimage.transform
import matplotlib.pyplot as plt

def load_data(directory):
    training_data=[]
    labels=[]
    images=[]
    for i in os.listdir(directory):
        if(i[0]=="0"):
            for j in os.listdir(directory+i):
                if j.endswith(".ppm"):
                    labels.append(int(i))
                    images.append(cv2.imread(directory+i+"/"+j))
    return labels,images
        
# Loading the training and testing dataset 
training_labels, training_images = load_data("Training/")
testing_labels, testing_images = load_data("Testing/")
print(set(training_labels))
```
