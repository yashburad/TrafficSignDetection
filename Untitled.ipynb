```python
import skimage
from PIL import Image
import os
import cv2
import skimage.transform
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

def load_data(directory):
    training_data=[]
    for i in os.listdir(directory):
        if(i[0]=="0"):
            for j in os.listdir(directory+i):
                if j.endswith(".ppm"):
                    training_data.append([cv2.imread(directory+i+"/"+j),int(i)])
    return training_data
        
# Loading the training and testing dataset 
training_data= load_data("Training/")
testing_data = load_data("Testing/")

```

```python
from sklearn.utils import shuffle

#shuffling training data
training_data=shuffle(training_data)
a=np.asarray(training_data)
training_labels, training_images = list(a[:,1]),np.array(list(a[:,0]))

#shuffling testing data
testing_data=shuffle(testing_data)
a=np.asarray(testing_data)
testing_labels, testing_images = list(a[:,1]),list(a[:,0])

print(set(training_labels),training_images[0].shape)
```

```python
def display_images(labels,images): #need to modify this code
    #Display the first image of each label.
    unique_labels = set(labels) #Grouping same labels together
    plt.figure(figsize=(15, 15)) #Sets the size of figure which will contain all images displayed to 15x15 inches
    i = 1
    for label in unique_labels:
        image = images[labels.index(label)]
        plt.subplot(8, 8, i)  # A grid of 8 rows x 8 columns in the 15x15 inch plot figure created above
        plt.axis('off')
        plt.title("Label {0} ({1})".format(label, labels.count(label)))
        i += 1
        _ = plt.imshow(image)
    plt.show()
display_images(training_labels,training_images)
```

```python
from skimage.transform import resize

resized=[]
for i in training_images:
    resized.append(resize(i,(64,64), mode ='constant'))
training_images=np.asarray(resized)

resized=[]
for i in testing_images:
    resized.append(resize(i,(64,64), mode ='constant'))
testing_images=np.asarray(resized)
```

```python
def bar_graph(labels):
    unique_labels = set(labels) #Grouping same labels together
    y_axis=[]
    for label in unique_labels:
        y_axis.append(labels.count(label))
    plt.figure(figsize=(12, 5))
    plt.bar(range(0, len(y_axis)), y_axis)
    plt.title("Distribution of the training dataset")
    plt.xlabel("Class number")
    plt.ylabel("Number of images")
    plt.show()
    
bar_graph(training_labels)
```

```python
#adding a depth of 1

X_train=training_images
training_images=training_images.reshape(training_images.shape[0],training_images.shape[1],training_images.shape[2],1)
print(X_train.shape,training_images.shape)
```

```python
from keras.preprocessing.image import ImageDataGenerator


image_dimensions = (64,64,3)

dataGen= ImageDataGenerator(width_shift_range=0.1,   # 0.1 = 10%     IF MORE THAN 1 E.G 10 THEN IT REFFERS TO NO. OF  PIXELS EG 10 PIXELS
                            height_shift_range=0.1,
                            zoom_range=0.2,  # 0.2 MEANS CAN GO FROM 0.8 TO 1.2
                            shear_range=0.1,  # MAGNITUDE OF SHEAR ANGLE
                            rotation_range=10)  # DEGREES
dataGen.fit(training_images)
batches= dataGen.flow(training_images,training_labels,batch_size=20)  # REQUESTING DATA GENRATOR TO GENERATE IMAGES  BATCH SIZE = NO. OF IMAGES CREAED EACH TIME ITS CALLED
X_batch,y_batch = next(batches)
 
# TO SHOW AGMENTED IMAGE SAMPLES
fig,axs=plt.subplots(1,15,figsize=(20,5))
fig.tight_layout()
 
for i in range(15):
    axs[i].imshow(X_batch[i].reshape(image_dimensions[0],image_dimensions[1]))
    axs[i].axis('off')
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
from keras.layers import Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
```

```python
training_images, X_validation, training_labels, y_validation = train_test_split(training_images, training_labels, test_size=0.2)
training_labels=to_categorical(training_labels)
y_validation=to_categorical(y_validation)
```

```python
def cnn_Model():
 
    size_of_filter=(5,5) # THIS IS THE KERNEL THAT MOVE AROUND THE IMAGE TO GET THE FEATURES.
                         # THIS WOULD REMOVE 2 PIXELS FROM EACH BORDER WHEN USING 32 32 IMAGE
   
    size_of_pool=(2,2)  # SCALE DOWN ALL FEATURE MAP TO GERNALIZE MORE, TO REDUCE OVERFITTING
    no_Of_Nodes = 500   # NO. OF NODES IN HIDDEN LAYERS
    model= Sequential()
    model.add((Conv2D(32,size_of_filter,input_shape=(image_dimensions[0],image_dimensions[1],1),activation='relu')))  # ADDING MORE CONVOLUTION LAYERS = LESS FEATURES BUT CAN CAUSE ACCURACY TO INCREASE
    model.add((Conv2D(32, size_of_filter, activation='relu')))
    model.add(MaxPooling2D(pool_size=size_of_pool)) # DOES NOT EFFECT THE DEPTH/NO OF FILTERS
 
    model.add((Conv2D(64,size_of_filter,activation='relu')))
    model.add((Conv2D(64, size_of_filter, activation='relu')))
    model.add(MaxPooling2D(pool_size=size_of_pool))
    model.add(Dropout(0.25))
 
    model.add(Flatten())
    model.add(Dense(no_Of_Nodes,activation='relu'))
    model.add(Dropout(0.5)) # INPUTS NODES TO DROP WITH EACH UPDATE 1 ALL 0 NONE
    model.add(Dense(len(training_labels),activation='softmax')) # OUTPUT LAYER
    # COMPILE MODEL
#     model.compile(Adam(lr=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
    model.compile(optimizer='adagrad', loss='categorical_crossentropy', metrics=['accuracy'])

    return model

model = cnn_Model()
print(model.summary())
history=model.fit(dataGen.flow(training_images,training_labels,batch_size=50),steps_per_epoch=2000,epochs=20,validation_data=(X_validation,y_validation),shuffle=1)
```
