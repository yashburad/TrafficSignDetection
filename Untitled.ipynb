```python
import skimage
from PIL import Image
import os
import cv2
import skimage.transform
import numpy as np
import matplotlib.pyplot as plt

def load_data(directory):
    training_data=[]
    for i in os.listdir(directory):
        if(i[0]=="0"):
            for j in os.listdir(directory+i):
                if j.endswith(".ppm"):
                    training_data.append([cv2.imread(directory+i+"/"+j),int(i)])
    return training_data
        
# Loading the training and testing dataset 
training_data= load_data("Training/")
testing_data = load_data("Testing/")
```

```python
from sklearn.utils import shuffle

#shuffling training data
training_data=shuffle(training_data)
a=np.asarray(training_data)
training_labels, training_images = list(a[:,1]),list(a[:,0])

#shuffling testing data
testing_data=shuffle(testing_data)
a=np.asarray(testing_data)
testing_labels, testing_images = list(a[:,1]),list(a[:,0])

print(set(training_labels))
```

```python
def display_images(images, labels):
    #Display the first image of each label.
    unique_labels = set(labels) #Grouping same labels together
    plt.figure(figsize=(15, 15)) #Sets the size of figure which will contain all images displayed to 15x15 inches
    i = 1
    for label in unique_labels:
        # Pick the first image for each label.
        image = images[labels.index(label)]
        plt.subplot(8, 8, i)  # A grid of 8 rows x 8 columns in the 15x15 inch plot figure created above
        plt.axis('off')
        plt.title("Label {0} ({1})".format(label, labels.count(label)))
        i += 1
        _ = plt.imshow(image)
    plt.show()
display_images(training_images, training_labels)
```
