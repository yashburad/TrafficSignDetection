**This project deals with building a crucial component of driverless cars viz. automatic traffic sign recognition. The model uses the Belgium TS dataset for this purpose.**

The dataset is split into training and testing data and has the following characteristics:

1.   The image size originally are inconsistent  but then reshaped to 64 (width) x 64 (height) x 3(RGB color channels).
2.   The training data set contains 4575 images with 62unique labels.
3.   The testing dataset contains 2520 images with 53 unique labels.

## **Procedure**

We import all the libraries that will be required for image augmentation , convolutional neural network, numpy array manipulation and converting images from RGB to gray.  

```python
import skimage
from PIL import Image
import os
import cv2
import skimage.transform
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from skimage.transform import resize
from keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
from keras.layers import Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
```

The data set is read into an array of array, as follows:

1. list training_data : a list of [image,image_label] where each image_label represents the label of the image.

2. list testing_data : a list of [image,image_label] where each image_label represents the label of the image.
    
We read the data from the "Training/" and "Testing/" directories and use them for the CNN.

```python
def load_data(directory):
    training_data=[]
    for i in os.listdir(directory):
        if(i[0]=="0"):
            for j in os.listdir(directory+i):
                if j.endswith(".ppm"):
                    training_data.append([cv2.imread(directory+i+"/"+j),int(i)])
    return training_data
        
# Loading the training and testing dataset 
training_data= load_data("Training/")
testing_data = load_data("Testing/")
```

### **Shuffling** ##

Here we shuffle the training and test data to remove biasness and make the dataset more evenly balanced. This help our machine to prevent overfitting. The lack of imbalance could cause our machine to predict the wrong label.

```python
training_data=shuffle(training_data)
a=np.asarray(training_data)
training_labels,training_images = np.array(list(a[:,1])),np.array(list(a[:,0]))

testing_data=shuffle(testing_data)
a=np.asarray(testing_data)
testing_lables,testing_images = np.array(list(a[:,1])),np.array(list(a[:,0]))
```

We see the distribution of the dataset with the labels in the below code and plot a bar graph of the training dataset. We find that the training dataset is highly imbalanced with Label 22 having 375 images and Label 11 having 7 images. We try to tackle this problem later onwards. 

```python
def bar_graph(labels):
    unique_labels = set(labels) #Grouping same labels together
    y=[]
    for label in unique_labels:
        y.append(labels.count(label))
    plt.figure(figsize=(16, 5))
    plt.bar(range(0, len(y)), y)
    plt.title("Distribution of the training dataset")
    plt.xlabel("Label number")
    plt.ylabel("Number of images")
    plt.show()
    
bar_graph(list(training_labels))
```

Since the images in the training dataset have different dimensions, we convert every image to 64x64 dimension as working on different dimension would be undesirble for our CNN

```python
resized=[]
for i in training_images:
    resized.append(resize(i,(64,64), mode ='constant'))
training_images=np.asarray(resized)

resized=[]
for i in testing_images:
    resized.append(resize(i,(64,64), mode ='constant'))
testing_images=np.asarray(resized)
```

```python
print(training_images.shape)
```

We just print one image from each label from the training dataset along with the number of images present for that label

```python
def display_images(labels,images,type): 
    setLabels = set(labels) 
    plt.figure(figsize=(15, 15))
    i = 1
    for label in setLabels:
        image = images[labels.index(label)]
        plt.subplot(7, 10, i)
        plt.axis('off')
        plt.title("Label "+str(label)+"("+str(labels.count(label))+")")
        if(type == "gray"):
            plt.imshow(image,cmap="gray")
        else:
            plt.imshow(image)
        i += 1
    plt.show()
    
display_images(list(training_labels),training_images,"normal")
```

### **Splitting** ###

Now we split it testing data into 80:20 ratio to testing and validation data.We will be using the validation dataset in our model learning and testing dataset to find the accuracy of the model on the testing dataset

```python
testing_images, validation_images, testing_labels, validation_labels = train_test_split(testing_images,testing_lables, test_size=0.2)

# print(testing_images.shape)
# print(testing_labels.shape)

# print(validation_images.shape)
# print(validation_labels.shape)
```

We convert our 3 channel (RGB) images (32 x 32 x 3) into a grayscale image (32 x 32 x1). We implement gray scaling of the images on all the datasets to check if grayscaling help us in removing noise from our dataset and make our CNN model more efficient

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

training_images=np.array(list(map(rgb2gray,training_images)))  
validation_images=np.array(list(map(rgb2gray,validation_images)))
testing_images=np.array(list(map(rgb2gray,testing_images)))
```

```python
display_images(list(training_labels),training_images,"gray") #display the converted images
```

```python
# Adding a depth of 1 to eack dataset to make dimension of every image to (64,64,1)

training_images=training_images.reshape(training_images.shape[0],training_images.shape[1],
                                        training_images.shape[2],1)
validation_images=validation_images.reshape(validation_images.shape[0],validation_images.shape[1],
                                            validation_images.shape[2],1)
testing_images=testing_images.reshape(testing_images.shape[0],testing_images.shape[1],
                                      testing_images.shape[2],1)
```

### **Data Augmentation**

Now, In the initial training dataset we can see that:

*   Label 22 has 375 images.
*   Label 11 has 7 images.

We see that the initial dataset has the problem of class imbalance. We try to solve this imbalance and give our dataset different types of input using **data augmentation** to make our model more robust.

We use the following image augmentation to try to make up for the imbalance:

1. Random Rotation between -100 and 100 angles
2. Changes Zoom_range and shear_range of the images and set it 0.25 and 0.15 respectively. 0.25 can imply 0.75 to 1.25.
3. Width_shift_range & Height_shift_range : Set to 0.15. Width_shift actually shift the image to the left or right(horizontal shifts) by 15% and height_shift vertically by 15%.

```python
imageDimensions = (64,64,1)

dataFit= ImageDataGenerator(rotation_range=10, # rotation degrees - -10 to 10
                            width_shift_range=0.15,   # 0.15 = 15%
                            height_shift_range=0.15,
                            zoom_range=0.25,  # 0.25 means it can go from 0.75 TO 1.25
                            shear_range=0.15) # value of the shear angle 

dataFit.fit(training_images) #this does data augmentation on our training dataset
```

We implement one hot encoding. A one hot encoding allows the representation of categorical data to be more expressive.

```python
classes=len(set(training_labels))
training_labels = to_categorical(training_labels,classes)
validation_labels = to_categorical(validation_labels,classes)
testing_labels = to_categorical(testing_labels,classes)

# print(len(set(training_labels)))
```

## Convolutional Neural Network ## 

We implement a model which has four layers of feature - 32,32 and 64,64 . All of the feature layer uses relu as the activation function. We keep the size of the filter at (3,3) throughout the modela and drops 25% of the nodes after hte firs two layers and then 50% after another two layers in our model during training to prevent overfitting, . We use adam optimizer to compile the model.

```python
def myModel():
    model= Sequential()
    model.add((Conv2D(32,(3,3),padding='same',input_shape=(imageDimensions[0],imageDimensions[1],1),activation='relu')))
    model.add((Conv2D(32, (3,3), activation='relu')))
    model.add(MaxPooling2D(pool_size=(2,2))) 
    model.add(Dropout(0.25))
 
    model.add((Conv2D(64, (3,3), padding='same',activation='relu')))
    model.add((Conv2D(64, (3,3), activation='relu')))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.5))
 
    model.add(Flatten())
    model.add(Dense(512,activation='relu'))
    model.add(Dropout(0.5)) 
    model.add(Dense(classes,activation='softmax')) 
    # COMPILE MODEL
    model.compile(Adam(lr=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
    return model
```

```python
batchSize=50
epochsVal=40
stepsPerEpoch=90

model = myModel()
print(model.summary())
history=model.fit_generator(dataFit.flow(training_images,training_labels,batch_size=batchSize),
                            steps_per_epoch=stepsPerEpoch,epochs=epochsVal,
                            validation_data=(validation_images,validation_labels),shuffle=1,verbose=1)
 
```

```python
plt.figure(1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['training','validation'])
plt.title('Accuracy')
plt.xlabel('epoch')
plt.figure(2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['training','validation'])
plt.title('Loss')
plt.xlabel('epoch')
plt.show()
score =model.evaluate(testing_images,testing_labels,verbose=0)
print('Test Score:',score[0])
print('Test Accuracy:',score[1])
```

```python
test_predictions = model.predict_classes(testing_images, batch_size=batchSize, verbose =1)
```

## Confusion Matrix ##

```python
from sklearn.metrics import confusion_matrix

values=[]
for i in range(testing_labels.shape[0]):
    for j in range(0,62):
        if(testing_labels[i][j]==1):
            values.append(j)

cm_test = confusion_matrix(values, test_predictions)
cm_test = cm_test.astype('float') / cm_test.sum(axis=1)[:, np.newaxis]  #To nomalize the confusion matrix
```

```python
import seaborn as sns
f = plt.figure(figsize=(40,20))
ax=f.add_subplot(1,1,1)
sns.heatmap(cm_test, linewidths=1, square = True, cmap = 'Blues')
ax.set_title('Confusion Matrix');
ax.set_ylabel('Correct label')
ax.set_xlabel('CNN Predicted label')
plt.show()
```
